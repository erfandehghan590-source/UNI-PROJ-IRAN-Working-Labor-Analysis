```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('gorgan_census95_activity.csv')
df
```

```python
df_age_15_64=df[df['age_group']!='10-14'][df[df['age_group']!='10-14']['age_group']!='65+']
df_age_15_64
```

```python
print(round(df_age_15_64.p_active_emp.sum()/df_age_15_64.p_total.sum()*100,3),"%")
```

```python
df_age_15_64_urban=df_age_15_64[df_age_15_64['settlement']=='شهري']
df_age_15_64_urban
```

```python
df_age_15_64_rural=df_age_15_64[df_age_15_64['settlement']=='روستايي']
df_age_15_64_rural
```

```python
urban_ratio_gorgan=round(df_age_15_64_urban.p_active_emp.sum()/df_age_15_64_urban.p_total.sum()*100,3)
print(urban_ratio_gorgan,"%")
```

```python
rural_ratio_gorgan=round(df_age_15_64_rural.p_active_emp.sum()/df_age_15_64_rural.p_total.sum()*100,3)
print(rural_ratio_gorgan,"%")
```

```python
print(round(rural_ratio_gorgan-urban_ratio_gorgan,3),'%')
```

```python
df_age_15_64_women=df_age_15_64[df_age_15_64['gender']=='زن']
df_age_15_64_women
```

```python
df_age_15_64_men=df_age_15_64[df_age_15_64['gender']=='مرد']
df_age_15_64_men
```

```python
men_ratio_gorgan=round(df_age_15_64_men.p_active_emp.sum()/df_age_15_64_men.p_total.sum()*100,3)
print(men_ratio_gorgan,"%")
```

```python
women_ratio_gorgan=round(df_age_15_64_women.p_active_emp.sum()/df_age_15_64_women.p_total.sum()*100,3)
print(women_ratio_gorgan,"%")
```

```python
print(round(men_ratio_gorgan-women_ratio_gorgan,3),'%')
```

```python
list_percentage_gorgan=[]
list_age_groups=[]
for i in df['age_group']:
    list_percentage_gorgan.append(round(df[df['age_group']==i].p_active_emp.sum()/df[df['age_group']==i].p_total.sum()*100,3))
    list_age_groups.append(i)
    if i == '65+':
        break
list_percentage_gorgan
```

```python
plt.figure(figsize=(15,9),dpi=150)
plt.bar(list_age_groups,list_percentage_gorgan)
plt.title('emp percentage by group',size='20')
plt.xlabel('age group',size='18')
plt.ylabel('employment percentage',size='18')
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.xticks(rotation=45)
plt.show()
```

```python
list_percentage_gorgan_urban=[]
list_age_groups=[]
for i in df['age_group']:
    list_percentage_gorgan_urban.append(round(df[df['settlement']=='شهري'][df[df['settlement']=='شهري']['age_group']==i].p_active_emp.sum()/df[df['settlement']=='شهري'][df[df['settlement']=='شهري']['age_group']==i].p_total.sum()*100,3))
    list_age_groups.append(i)
    if i == '65+':
        break
list_percentage_gorgan_urban
```

```python
list_percentage_gorgan_rural=[]
list_age_groups=[]
for i in df['age_group']:
    list_percentage_gorgan_rural.append(round(df[df['settlement']=='روستايي'][df[df['settlement']=='روستايي']['age_group']==i].p_active_emp.sum()/df[df['settlement']=='روستايي'][df[df['settlement']=='روستايي']['age_group']==i].p_total.sum()*100,3))
    list_age_groups.append(i)
    if i == '65+':
        break
list_percentage_gorgan_rural
```

```python
plt.figure(figsize=(15,9),dpi=150)
plt.plot(list_age_groups,list_percentage_gorgan_urban,color='#c92525',label='urban')
plt.plot(list_age_groups,list_percentage_gorgan_rural,color='#43802b',label='rural')
plt.legend(fontsize='15')
plt.title('emp percentage by group',size='20')
plt.xlabel('age group',size='18')
plt.ylabel('employment percentage',size='18')
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.xticks(rotation=45)
plt.show()
```

```python
list_percentage_gorgan_rural_men=[]
list_percentage_gorgan_rural_women=[]
list_age_groups=[]
for i in range(len(df['age_group'])):
    if df.gender.iloc[i]=='مرد':
        list_percentage_gorgan_rural_men.append(round(df[df['settlement']=='روستايي'][df[df['settlement']=='روستايي']['age_group']==df['age_group'][i]].p_active_emp.sum()/df[df['settlement']=='روستايي'][df[df['settlement']=='روستايي']['age_group']==df['age_group'][i]].p_total.sum()*100,3))
        
    
    elif df.gender.iloc[i]=='زن':
        list_percentage_gorgan_rural_women.append(round(df[df['settlement']=='روستايي'][df[df['settlement']=='روستايي']['age_group']==df['age_group'][i]].p_active_emp.sum()/df[df['settlement']=='روستايي'][df[df['settlement']=='روستايي']['age_group']==df['age_group'][i]].p_total.sum()*100,3))
        
    list_age_groups.append(df['age_group'][i])    
    if df['age_group'][i] == '65+':
        break
list_percentage_gorgan_rural_men
```

```python

df_male_g=df[df['gender']=='مرد']
df_male_urban_g= df_male_g[df_male_g['settlement'] == 'شهري']
df_male_rural_g= df_male_g[df_male_g['settlement'] == 'روستايي']

df_female_g=df[df['gender']== 'زن']
df_female_urban_g= df_female_g[df_female_g['settlement'] == 'شهري']
df_female_rural_g= df_female_g[df_female_g['settlement'] == 'روستايي']

df_female_rural_g
```
